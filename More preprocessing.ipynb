**1. Normalization and Standardization**

Q1: Normalize the dataset using Min-Max normalization.

**2. Encoding Categorical Variables**

**3. Discretization and Binning**

Q7: Bin the Max AQI column into 4 categories: "Low", "Moderate", "High", and "Very High"
using equal-width binning. How many rows fall into each category?

```python
bins = 4
labels = ["Low", "Moderate", "High", "Very High"]

df['AQI_Category'] = pd.cut(df['Max AQI'], bins=bins, labels=labels)

category_counts = df['AQI_Category'].value_counts()
print(category_counts)
```

**4. Handling Skewed Data**

Q8: Check the skewness of the Max AQI and Median AQI columns. Which of these columns is
highly skewed?

```python
max_aqi_skew = df['Max AQI'].skew()
median_aqi_skew = df['Median AQI'].skew()

print(f"Skewness of Max AQI: {max_aqi_skew}")
print(f"Skewness of Median AQI: {median_aqi_skew}")

if abs(max_aqi_skew) > abs(median_aqi_skew):
    print("Max AQI more skewed.")
else:
    print("Median AQI more skewed.")
```

Q9: Apply a log transformation to the Max AQI column to handle its skewness. How does this
affect the skewness?

```python
import numpy as np

df['Log_Max_AQI'] = np.log(df['Max AQI'] + 1)

original_skewness = df['Max AQI'].skew()
log_skewness = df['Log_Max_AQI'].skew()

print(f"Skewness of Max AQI (Original): {original_skewness}")
print(f"Skewness of Max AQI (Log Transformed): {log_skewness}")
```

Q10: After applying the log transformation, what changes do you observe in the distribution of
the Max AQI data?

- After the log transformation, the skewness of Max AQI dropped from 17.55 to 0.51. This makes the data more evenly distributed

**6. Multicollinearity**

Q12: Calculate the correlation matrix for all numeric variables. Which variables are highly
correlated with each other?

```python
df_encoded = pd.get_dummies(df, columns=['State', 'County'])

correlation_matrix = df_encoded.corr()

print(correlation_matrix)

highly_correlated_pairs = correlation_matrix[(correlation_matrix > 0.8) & (correlation_matrix != 1)]

print("Highly correlated pairs:")
print(highly_correlated_pairs)
```

Q13: Based on the correlation matrix, identify pairs of features that might cause multicollinearity.
How could you address this issue?

```python
threshold = 0.8

correlated_features = []
for column in correlation_matrix.columns:
    for row in correlation_matrix.index:
        if abs(correlation_matrix.loc[row, column]) > threshold and row != column:
            correlated_features.append((row, column))

correlated_features = list(set(tuple(sorted(pair)) for pair in correlated_features))

print("Highly correlated feature pairs:")
for pair in correlated_features:
    print(f"{pair[0]} and {pair[1]}: Correlation = {correlation_matrix.loc[pair[0], pair[1]]}")
```

- We can address this by removing one of the correlated features like Median AQI or 90th Percentile AQI
- We can also use regularization methods like Ridge or Lasso regression to handle multicollinearity without dropping features

Q14: Remove one of the highly correlated features from the dataset. Recompute the correlation
matrix and compare it with the original one. What changes do you observe?

```python
df_dropped = df_encoded.drop(columns=['Median AQI'])

new_corr_matrix = df_dropped.corr()

print("\nChanges in the correlation matrix after removing 'Median AQI':")
difference = correlation_matrix.subtract(new_corr_matrix, fill_value=0)

difference = difference[difference != 0].dropna(how='all').dropna(axis=1, how='all')

print(difference)
```
